```python
from nltk.corpus import stopwords
```

```python
sw=set(stopwords.words('english'))
```

```python
import string
punctuations=string.punctuation
```

```python
sw.update(punctuations)
```

```python
def useful_words(word,sw):
    for i in range(len(word)):
        word[i]=word[i].lower()
    useful_word=[w for w in word if w not in sw]
    return useful_word
```

```python
u_words=[]
for i in range(len(data)):
    u_words.append(useful_words(data[i],sw))
```

```python
from nltk import pos_tag
```

```python
from nltk.corpus import wordnet
```

```python
def get_simple_pos(word):
    if word.startswith("J"):
        return wordnet.ADJ
    elif word.startswith("V"):
        return wordnet.VERB
    elif word.startswith("N"):
        return wordnet.NOUN
    elif word.startswith("R"):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def clean_review(sentence):
    clean_words=[]
    for i in range(len(sentence)):
        pos=pos_tag([sentence[i]])
        clean_words.append(lemmatizer.lemmatize(sentence[i],get_simple_pos(pos[0][1])))
    return clean_words
```

```python
for i in range(len(u_words)):
    u_words[i]=clean_review(u_words[i])
```

```python
a="harsh"
pos=pos_tag([a])
print(get_simple_pos(pos[0][1]))
```

```python
corpus=[]
for i in range(len(u_words)):
    corpus.append(' '.join(u_words[i]))
```
